```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
#converting the sklearn dataset into the data frame here
```

```python
#cancer['target']#here our target column contain malignant=0 and benign=1
#here we are not going to consider it in the PCA analysis just keep in mind
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
```

```python
scaled_data.min()
```

```python
scaled_data.max()
```

```python
from sklearn.decomposition import PCA
    
```

```python
pca=PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'])
plt.xlabel('First principal component')
plt.ylabel("2 nd PCA")
```

```python
pca.components_
```

```python
df_comp=pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
#this graph shows us the relation between the pca component and various featured higher the number
#which is highly correlated
```
