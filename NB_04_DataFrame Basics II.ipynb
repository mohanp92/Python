## Filtering DataFrames with many Conditions (AND)

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head(10)
```

```python
mask1 = titanic.sex == "male"
mask1.head()
```

```python
mask2 = titanic.age > 14
mask2.head()
```

```python
(mask1 & mask2).head()
```

```python
male_adult = titanic.loc[mask1 & mask2, ["survived", "pclass", "sex", "age"]]
male_adult.head(20)
```

```python
male_adult.info()
```

```python
male_adult.describe()
```

```python
titanic.describe()
```

## Advanced Filtering with between(), isin() and ~

```python
import pandas as pd
```

```python
summer = pd.read_csv("summer.csv")
```

```python
summer.head()
```

```python
og_1988 = summer.loc[summer.Year == 1988]
```

```python
og_1988.head()
```

```python
og_1988.tail()
```

```python
og_1988.info()
```

```python
og_since1992 = summer.loc[summer.Year >= 1992]
```

```python
og_since1992.head()
```

```python
og_since1992.tail()
```

```python
summer.Year.between(1960, 1969).head()
```

```python
og_60s = summer.loc[summer.Year.between(1960, 1969, inclusive=True)]
#inclusive true means it include the boundries here that is extream ends
```

```python
og_60s.head()
```

```python
og_60s.tail()
```

```python
my_favourite_games = [1972, 1996]
```

```python
summer.Year.isin(my_favourite_games).head()
#just check it out whether the year 1972 and year 1996 are available then apply loc to get the 
#remaining information here
```

```python
og_72_96 = summer.loc[summer.Year.isin(my_favourite_games)]
```

```python
og_72_96.head()
```

```python
og_72_96.tail()
```

```python
og_not_72_96 = summer.loc[~summer.Year.isin(my_favourite_games)]
```

```python
og_not_72_96.head()
```

```python
og_not_72_96.Year.unique()
```

## any() and all()

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.sex == "male"
```

```python
(titanic.sex == "male").any()
# whether any element is true or not . i.e it is checking whether any element is match out of all
```

```python
(titanic.sex == "male").all()
#all element is male or not
```

```python
titanic_male=titanic.loc[titanic.sex=='male']
```

```python
(titanic_male.sex=='female').any()
```

```python
(titanic_male.sex=='male').all() # in this df all gender is male 
```

```python
(titanic.age == 80.0).any()
```

```python
pd.Series([-1, 0.5 , 1, -0.1, 0]).any()
```
