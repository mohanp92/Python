### Indexing Operator iloc (location based indexing) 

```python
medals = pd.read_csv("summer.csv")
```

```python
medals.head()
```

```python
medals.tail()
```

```python
medals.info()
```

#### Selecting Rows with iloc

```python
medals.iloc[0] # iloc is the postion based indexing and loc is the label based indexing 
#so iloc only take the int values
```

```python
type(medals.iloc[0])
```

```python
medals.iloc[1]
```

```python
medals.iloc[-1] # it will give the last rows here
```

```python
medals.iloc[:5] #It will show you rows from the 0 to 4 . Note 5 won't show here
#when you do the same thing with loc then it also show you the 5th row as well
```

```python
medals.iloc[-5:]# last 5
```

```python
medals.iloc[456:459] # shows you from 456 to 459
```

```python
medals.iloc[[2,45,765]] # gives you 2,45,765
```

```python
medals.iloc[[5,56,67,789]]# pass the list of rows then get the out put
```

#### Slicing Rows and Columns with iloc

```python
medals.head()
```

```python
medals.iloc[0,4]
#as we seen previously 1 st position was for the row now it's time for the colmn
#it will show you 0 th row 4 th column value 
```

```python
medals.iloc[0,0:3]# 0th row and from 0 to 2 scond column values
```

```python
medals.iloc[0,[0,2,6,8]]# 0 th row and 0, 2,6,8 column value
```

```python
medals.iloc[34:39,[0,2,6,8]] # 34 to 38 rows and same column values
```

#### Selecting Columns with iloc

```python
medals.iloc[:, 4].equals(medals.Athlete)
```

```python
medals["Athlete"]
```

### Index Operator loc (label based indexing)

```python
medals = pd.read_csv("summer.csv", index_col="Athlete")
#Here created a data frame with column as an index here
```

```python
medals.head()
```

#### Selecting Rows with loc

```python
medals.iloc[2]# still postion based indexing is working here
```

```python
medals.loc["DRIVAS, Dimitrios"]# we directly pass the label of index 
```

```python
medals.loc["PHELPS, Michael"]# label indexing also use full when you want the data of perticular
#customer her phelps michel
```

```python
medals.loc["PHELPS, Michael"].iloc[0] # Out of many result we get the 0 i.e first row here
```

#### Slicing Rows and Columns with loc

```python
medals.loc["PHELPS, Michael", "Medal"]# same as iloc here just passing the column name here
```

```python
medals.loc["PHELPS, Michael", ["Event","Medal"]]#Select 2 columns
```

```python
medals.loc[["PHELPS, Michael", "LEWIS, Carl"], ["Event","Medal"]] # Selecting the 2 label from the 
# index and select 2 columns from the list
```

```python
medals.head(10)
```

```python
medals.loc[:"CHASAPIS, Spiridon"]# selecting the label until the "CHASAPIS," 
#In label based indexing both starting point and ecding point also included
```

```python
#medals.loc[:"PHELPS, Michael"] for label based indexing label should be unique i.e that only appear
# one time
```

```python
#medals.loc["HAJOS, Alfred":]
```

```python
medals.head(20)
```

```python
medals.loc["DRIVAS, Dimitrios":"BLAKE, Arthur"]
```
