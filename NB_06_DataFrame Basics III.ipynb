## Ranking DataFrames with rank()

```python
import pandas as pd
```

```python
sales = pd.Series([15, 32, 45, 21, 55, 15, 0],  index = ["Mo", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
```

```python
sales
```

```python
sales.sort_values(ascending = False)
```

```python
sales.rank(ascending=False, method = "min").sort_values(ascending = True)
```

```python
sales.rank(ascending=False, method = "min", pct=True).sort_values()
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.fare.rank(ascending = False)
```

```python
titanic["fare_rank"] = titanic.fare.rank(ascending = False, method="min")
```

```python
titanic.head()
```

```python
titanic.sort_values("fare", ascending= False)
```

```python
titanic.drop(columns = "fare_rank", inplace= True)
```

## apply(), map() and applymap()

```python
import pandas as pd
```

```python
sales = pd.read_csv("sales.csv", index_col = 0)
```

```python
sales
```

```python
sales.info()
```

```python
sales.min(axis =0)
```

```python
sales.min(axis = 1)
```

```python
def range(series):
    return series.max() - series.min()
```

```python
sales.apply(lambda x: x.max() - x.min(), axis = 0)
#lambda jsut telling that we are creating function and x means we are taking the parameter 
#instead of writting the user define fuuction we use lambda  fucntion here
```

```python
summer = pd.read_csv("summer.csv")
```

```python
summer.head()
```

```python
summer.Athlete.apply(lambda x: x[0])
#here we creadted the user defined function which take x and return the x[0] value here 
```

```python
summer.Athlete[0][0]
```

```python
summer.Athlete.map(lambda x: x[0])
# here i don't thik there is any difference between apply, map, and applymap
#apply works on a row / column basis of a DataFrame, applymap works element-wise on a DataFrame,
#and map works element-wise on a Series.
```

```python
summer.iloc[:,1:3].applymap(lambda x: x[0])
```

```python
sales.applymap(lambda x: 0.4*x-5)
```

```python
sales*0.4-5
```

## Hierarchical Indexing (MultiIndex) Intro
